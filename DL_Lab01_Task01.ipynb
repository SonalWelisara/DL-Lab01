Create a 4x4 array from an exponential distribution

```python
exponential_array = np.random.exponential(size=(4, 4))
print(exponential_array)
```

Creating random arrays with exponential, nomral and uniform distribution


```python
rand_array = np.random.rand(100000)
randn_array = np.random.randn(100000)
exp_array = np.random.exponential(size=100000)
```

visualizing the distribution histogram

```python
plt.hist(rand_array, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(randn_array, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.hist(exp_array, density=True, bins=100, histtype="step", color="green", label="Exponential")

plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Create the X and Y grid and calculate Z


```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

Create a figure and axes for 3D plotting

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

Downloading and reading Pokemon data set


```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin1')
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
pokemon_stats = df[features]
```

```python
pearson = pokemon_stats.corr(method='pearson')
spearman = pokemon_stats.corr(method='spearman')
```

Visualizing the Heatmaps

```python
plt.figure(figsize=(10, 8))
sns.heatmap(pearson, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation Heatmap')
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(spearman, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Rank Correlation Heatmap')
plt.show()
```
